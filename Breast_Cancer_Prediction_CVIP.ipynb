**INTERQUARTILE FORMULA TO REMOVE OUTLIERS**

```python
import numpy as np
for col in df.columns:
  a=df[col].quantile(0.25)
  b=df[col].quantile(0.75)
  iqr=b-a
  upper_limit=b+(1.5*iqr)
  lower_limit=a-(1.5*iqr)
  df[col] = np.where(df[col] < lower_limit, lower_limit,
  np.where(df[col] > upper_limit, upper_limit,df[col]))
```

**BOXPLOT FOR REMOVED OUTLIERS**

**SPLITTING X AND Y VALUES**

```python
x=df.drop(['diagnosis'],axis=1)
y=df.diagnosis
```

**IMPORTING TRAIN TEST AND SPLIT**

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

**IMPORTING ACCURACY SCORE**

```python
from sklearn.metrics import accuracy_score
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(x,y,test_size=0.5,random_state=42)
```

```python
a=StandardScaler()
X_train=a.fit_transform(X_train)
X_test = a.transform(X_test)
```

**IMPORTING NECESSARY ALGORITHMS**

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
```

**DEFINING ALGORITHMS**

```python
algorithm = ['DecisiontreeClassifier','RandomForestClassifier','AdaBoostClassifier','SVC','GaussianNB','KNeighborsClassifier','LogisticRegression']
value = [DecisionTreeClassifier(criterion="entropy",random_state=42),RandomForestClassifier(n_estimators=300),AdaBoostClassifier(n_estimators=100),SVC(kernel="rbf",random_state=15),GaussianNB(),KNeighborsClassifier(n_neighbors=9,weights='uniform'),LogisticRegression()]
models = dict(zip(algorithm,value))
models
```

**DEFINING PREDICTED VALUE AND ACCURACY SCORE**

```python
predicted =[]
for name,algo in models.items():
    model=algo
    model.fit(X_train,Y_train)
    predict = model.predict(X_test)
    accuracy = accuracy_score(Y_test, predict)*100
    predicted.append(accuracy)
    print(name,accuracy)
```

**PLOTTING ACCURACY SCORES OF THE DEFINED ALGORITHMS**

```python
plt.figure(figsize = (15,15))
sns.barplot(x = predicted, y = algorithm, palette='magma')
plt.title("ACCURACY VALUES", fontsize=20)
```
